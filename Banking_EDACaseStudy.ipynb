## Importing Libraries

```python
#Importing libraries
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
query = 'SELECT * FROM Customers'
```

```python
#Show top 5 rows of the Customers table
df = pd.read_sql(query, conn)
df.head(5)
```

## Exploratory Data Analysis

```python
# Check the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Get a concise summary of the DataFrame
print("\nDataFrame Info:")
df.info()
```

```python
# Generate descriptive statistics for the dataframe
df.describe()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated_Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar', figsize=(8, 5))
plt.title('Income Band Distribution')
plt.xlabel('Income Band')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

## Data Cleaning and Preparation

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values[missing_values > 0])
if missing_values.sum() == 0:
    print("No missing values detected!")
```

```python
# Convert 'Joined Bank' to datetime format
df['Joined_Bank'] = pd.to_datetime(df['Joined_Bank'], format='%Y-%m-%d')
print("Date conversion completed successfully!")
print(f"Date range: {df['Joined_Bank'].min()} to {df['Joined_Bank'].max()}")
```

## Categorical Analysis

```python
# Analyze distribution of categorical columns
categorical_cols = ['BRId', 'GenderId', 'IAId', 'Amount_of_Credit_Cards', 'Nationality', 
                    'Occupation', 'Fee_Structure', 'Loyalty_Classification', 'Properties_Owned', 
                    'Risk_Weighting', 'Income Band']

for col in categorical_cols:
    print(f"\n{'='*50}")
    print(f"Value Counts for '{col}':")
    print(f"{'='*50}")
    print(df[col].value_counts())
```

### Univariate Analysis - Categorical Variables

```python
categorical_cols = ['BRId', 'GenderId', 'IAId', 'Amount_of_Credit_Cards', 'Nationality', 
                    'Occupation', 'Fee_Structure', 'Loyalty_Classification', 'Properties_Owned', 
                    'Risk_Weighting', 'Income Band']

fig = plt.figure(figsize=(20, 12))
for i, predictor in enumerate(categorical_cols):
    plt.subplot(3, 4, i + 1)
    sns.countplot(data=df, x=predictor, palette='Set2')
    plt.title(f'{predictor}', fontsize=11, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    
plt.tight_layout()
plt.show()
```

## Numerical Analysis

```python
numerical_cols = ['Age', 'Estimated_Income', 'Superannuation_Savings', 'Credit_Card_Balance', 
                  'Bank_Loans', 'Bank_Deposits', 'Checking_Accounts', 'Saving_Accounts', 
                  'Foreign_Currency_Account', 'Business_Lending', 'Properties_Owned']

# Create histograms with KDE for numerical columns
plt.figure(figsize=(18, 12))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 4, i + 1)
    sns.histplot(df[col], kde=True, color='steelblue', bins=30)
    plt.title(col, fontsize=11, fontweight='bold')
    plt.xlabel('')
    
plt.tight_layout()
plt.show()
```

## Correlation Analysis

```python
# Calculate correlation matrix for numerical columns
correlation_matrix = df[numerical_cols].corr()

# Create heatmap
plt.figure(figsize=(14, 11))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", 
            center=0, square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Numerical Features', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

## Bivariate Analysis

```python
# Bivariate analysis: Categorical variables with Nationality hue
categorical_cols_subset = ['BRId', 'GenderId', 'IAId', 'Amount_of_Credit_Cards', 
                           'Occupation', 'Fee_Structure', 'Loyalty_Classification', 'Income Band']

fig = plt.figure(figsize=(20, 10))
for i, predictor in enumerate(categorical_cols_subset):
    plt.subplot(2, 4, i + 1)
    sns.countplot(data=df, x=predictor, hue='Nationality', palette='husl')
    plt.title(f'{predictor} by Nationality', fontsize=10, fontweight='bold')
    plt.xticks(rotation=45, ha='right', fontsize=8)
    if i == 0:
        plt.legend(fontsize=8, loc='best')
    else:
        plt.legend([],[], frameon=False)
    
plt.tight_layout()
plt.show()
```

### Relationship Analysis - Regression Plots

```python
# Key variable pairs to analyze
pairs_to_plot = [
    ('Bank_Deposits', 'Saving_Accounts'),
    ('Checking_Accounts', 'Saving_Accounts'),
    ('Checking_Accounts', 'Foreign_Currency_Account'),
    ('Age', 'Superannuation_Savings'),
    ('Estimated_Income', 'Checking_Accounts'),
    ('Bank_Loans', 'Credit_Card_Balance'),
    ('Business_Lending', 'Bank_Loans'),
    ('Estimated_Income', 'Bank_Deposits')
]

fig = plt.figure(figsize=(16, 12))
for idx, (x_col, y_col) in enumerate(pairs_to_plot):
    plt.subplot(2, 4, idx + 1)
    sns.regplot(data=df, x=x_col, y=y_col, scatter_kws={'alpha': 0.3, 's': 20}, 
                line_kws={'color': 'red', 'linewidth': 2})
    plt.title(f'{x_col} vs {y_col}', fontsize=10, fontweight='bold')
    plt.xlabel(x_col, fontsize=9)
    plt.ylabel(y_col, fontsize=9)
    
plt.tight_layout()
plt.show()
```

## Key Insights from Exploratory Data Analysis

### 📊 **Deposits and Savings Behavior**
- Strong positive correlation exists between Bank Deposits and Saving Accounts
- This suggests that customers who actively deposit funds also tend to maintain substantial savings balances
- These variables may represent overlapping financial behaviors or complementary account management strategies

### 💰 **Income, Age, and Financial Accumulation**
- Moderate correlations observed between Age, Estimated Income, and various account balances
- Demonstrates a typical financial lifecycle: higher income earners and older individuals accumulate more savings and retirement funds
- Superannuation Savings show positive correlation with Age, reflecting retirement planning patterns

### 🏠 **Property Ownership Patterns**
- Property ownership shows weak correlations with banking variables
- This indicates that property ownership depends on external factors (real estate market, location, inheritance) not captured by banking metrics alone
- Property ownership appears to be independent of customer banking behavior

### 💳 **Business vs. Personal Banking**
- Business Lending shows moderate correlation with Bank Loans, suggesting some customers maintain both personal and business debts
- Business lending is relatively uncorrelated with deposit and property metrics, indicating it serves a distinct customer segment
- This represents a specialized banking need separate from personal wealth accumulation

### 👥 **Nationality and Customer Segments**
- Distribution analysis reveals diverse customer base across multiple nationalities
- Bivariate analysis shows nationality influences financial behaviors differently across customer segments
- Different nationalities may have distinct banking preferences and risk profiles

### 📈 **Income Band Insights**
- Customers are distributed across Low, Medium, and High income bands based on Estimated Income
- Income band serves as an important segmentation variable for targeting and risk assessment
- Strong inverse relationship expected between income band and credit-dependent behaviors

### ✅ **Data Quality**
- No missing values detected in the dataset
- All columns properly formatted and ready for further analysis and modeling
- Date fields successfully converted to datetime format for temporal analysis
